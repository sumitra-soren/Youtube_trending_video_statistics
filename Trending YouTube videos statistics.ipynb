## Data Cleaning

```python
youtube_data_copy = youtube_data.copy()
youtube_data_copy.head()
```

```python
del_cols = ['thumbnail_link','comments_disabled','ratings_disabled', 'video_error_or_removed' , 'description']
youtube_data_copy = youtube_data_copy.drop(del_cols, axis=1)
```

```python
youtube_data_copy.columns
```

```python
youtube_data_copy = youtube_data_copy.drop(['trend_duration'], axis = 1)
```

```python
youtube_data_copy['video_id'] = youtube_data_copy['video_id'].astype('str')
```

```python
youtube_data_copy['title'] = youtube_data_copy['title'].astype('str')
```

```python
youtube_data_copy['channel_title'] = youtube_data_copy['channel_title'].astype('str')
```

```python
youtube_data_copy['category_id'] = youtube_data_copy['category_id'].astype('str')
```

```python
youtube_data_copy['tags'] = youtube_data_copy['tags'].astype('str')
```

```python
youtube_data_copy['trending_date'] = youtube_data_copy['trending_date'].astype('str')
```

```python
date_pieces = youtube_data_copy['trending_date'].str.split('.')
```

```python
youtube_data_copy['Year'] = date_pieces.str[0].astype('int')
youtube_data_copy['Day'] = date_pieces.str[1].astype('int')
youtube_data_copy['Month'] = date_pieces.str[2].astype('int')

updatedyear = []
for i in range(len(youtube_data_copy)):
    y = youtube_data_copy.loc[i,"Year"]
    newy = y+2000
    updatedyear.append(newy)
    
for i in range(len(youtube_data_copy)):
    newy = updatedyear[i]
    tr = youtube_data_copy.loc[i,"Year"]
    youtube_data_copy['Year'].replace(to_replace = tr, value = newy, inplace = True)

del youtube_data_copy['trending_date']
youtube_data_copy['trending_date'] = pd.to_datetime(youtube_data_copy[['Year','Month','Day']], format = "%Y-%m-%d")
del youtube_data_copy['Year']
del youtube_data_copy['Day']
del youtube_data_copy['Month']
```

```python
youtube_data_copy['publish_time'] = pd.to_datetime(youtube_data_copy['publish_time'],
                                                   errors='coerce', format='%Y-%m-%dT%H:%M:%S.%fZ')
```

```python
youtube_data_copy.insert(4, 'publish_date', youtube_data_copy['publish_time'].dt.date)
youtube_data_copy['publish_time'] = youtube_data_copy['publish_time'].dt.time
```

```python
youtube_data_copy['publish_date'] = pd.to_datetime(youtube_data_copy['publish_date'], format = "%Y-%m-%d")
```

## Exploratory Data Analysis

```python
#top 5 trending videos
trending_videos = youtube_data_copy.reset_index().sort_values(by = ['views'], ascending = False)
trending_videos.drop_duplicates(subset = "video_id", keep = 'first', inplace = True)
trending_videos.set_index('video_id', inplace = True)
trending_videos.head(5)[['title', 'channel_title', 'category', 'views', 'likes']]
```

```python
#top 5 most liked videos
most_liked_videos = youtube_data_copy.reset_index().sort_values(by = ['likes'], ascending = False)
most_liked_videos.drop_duplicates(subset = "video_id", keep = 'first', inplace = True)
most_liked_videos.set_index('video_id', inplace = True)
most_liked_videos.head(5)[['title','channel_title','category', 'views', 'likes']]
```

```python
#top5 most viewed category
most_viewed_category = youtube_data_copy
most_viewed_category = most_viewed_category.groupby('category')['views'].agg('sum')
most_viewed_category = most_viewed_category.sort_values(ascending = False)
most_viewed_category.head()
```

```python
#top5 most liked category
most_liked_category = youtube_data_copy
most_liked_category = most_liked_category.groupby('category')['likes'].agg('sum')
most_liked_category = most_liked_category.sort_values(ascending = False)
most_liked_category.head()
```

```python
#top5 most disliked categories
most_disliked_category = youtube_data_copy
most_disliked_category = most_disliked_category.groupby('category')['dislikes'].agg('sum')
most_disliked_category = most_disliked_category.sort_values(ascending = False)
most_disliked_category.head()
```

```python
#ratio of likes and dislikes in different categories
likesdf = youtube_data_copy.groupby('category')['likes'].agg('sum')
dislikesdf = youtube_data_copy.groupby('category')['dislikes'].agg('sum')

ratiodf = likesdf/dislikesdf
ratiodf = ratiodf.sort_values(ascending = False)
ratiodf
```

```python
#top5 channels with most trending videos
youtube_data_copy['channel_title'].value_counts().head()
```

```python
#Maximum number of days to trending status for a video
timespan = youtube_data_copy
timespan['trend_duration'] = (timespan['trending_date'] - timespan['publish_date']).dt.days
timespan = timespan.sort_values(by = 'trend_duration', ascending = False)
timespan.head(5)[['category','title','trend_duration']]
```
